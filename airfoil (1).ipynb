# Hypertuning

```python
from sklearn.model_selection import GridSearchCV

# Define the hyperparameters and their possible values
param_grid = {
    'n_estimators': [100, 200, 300,400,50,20,50],
    'max_features': ['auto', 'sqrt','log2'],
    'max_depth': [None, 10, 20,5,15],
    'min_samples_split': [2,4,6,8, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create the RandomForest Regressor
rf = RandomForestRegressor()

# Perform Grid Search with 5-fold cross-validation
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,
                           cv=5, n_jobs=-1, verbose=2)

# Fit the Grid Search to your data
grid_search.fit(xtrain,ytrain)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

ybesttestpredr=grid_search.predict(xtest)
```

```python
sbscore=r2_score(ytest, ybesttestpredr)
print(sbscore)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from scipy.stats import randint

param_dist = {
    'n_estimators': randint(10, 200),
    'max_depth': randint(1, 20),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 20)
}

# Create your classifier
rf = RandomForestRegressor()

# Create the RandomizedSearchCV object
random_search = RandomizedSearchCV(
    rf, param_distributions=param_dist, n_iter=10, cv=5, n_jobs=-1, verbose=1, random_state=42
)

# Fit the model
random_search.fit(xtrain,ytrain)

# Get the best parameters and estimator
best_params = random_search.best_params_
best_estimator = random_search.best_estimator_

# Use the best estimator for predictions
y_pred = best_estimator.predict(xtest)
```

```python
srscore=r2_score(ytest, y_pred)
print(srscore)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv(r'C:\Users\boina\OneDrive\Desktop\sb\BizLand\AirfoilSelfNoise.csv')
```

```python
features = ['alpha','U_infinity']
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[features])
```

```python
pca = PCA(n_components=1)  
df_pca = pca.fit_transform(df_scaled)
```

```python
df_combined = pd.DataFrame(data=df_pca, columns=['PCA_Component'])
```

```python
df = pd.concat([df, df_combined], axis=1).drop(columns=features)
```

```python
df.head()
```

```python
x=df.drop(['SSPL'],axis=1)
Y=df['SSPL']
```
